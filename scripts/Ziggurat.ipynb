```python
import numpy as np
import seaborn as sns
from scipy.stats import norm, invnorm
import matplotlib.pyplot as plt
from matplotlib.collections import PatchCollection
from matplotlib.patches import Rectangle
```

```python
sns.set_style("whitegrid")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(14,10))
x = np.linspace(0,3.5, 100)
r1 = Rectangle((0, 0), -norm.ppf(0.02), 0.05)
r2 = Rectangle((0, 0.05), -norm.ppf(0.02), 0.05)
r3 = Rectangle((0, 0.10), -norm.ppf(0.05), 0.05)
r4 = Rectangle((0, 0.15), -norm.ppf(0.08), 0.05)
r5 = Rectangle((0, 0.20), -norm.ppf(0.12), 0.05)
r6 = Rectangle((0, 0.25), -norm.ppf(0.17), 0.05)
r7 = Rectangle((0, 0.30), -norm.ppf(0.23), 0.05)
r8 = Rectangle((0, 0.35), -norm.ppf(0.30), 0.05)
pc = PatchCollection([r1, r2, r3, r4, r5, r6, r7, r8], facecolor='white', alpha=1,
                         edgecolor='black')
ax.add_collection(pc)
ax.plot(x, norm.pdf(x),label='halfnorm pdf')
ax.set_xticks([-norm.ppf(0.30),-norm.ppf(0.23),-norm.ppf(0.17),-norm.ppf(0.12),-norm.ppf(0.08),-norm.ppf(0.05),-norm.ppf(0.02)])
ax.set_xticklabels(['$x[7]$','$x[6]$','$x[5]$', '$x[4]$', '$x[3]$', '$x[2]$', '$x[1],x[0]$'])
ax.set_yticks([0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45])
ax.set_yticklabels(['$y[0]$','$y[1]$', '$y[2]$', '$y[3]$', '$y[4]$', '$y[5]$', '$y[6]$', '$y[7]$'])

fig.savefig("ziggurat.pgf", format='pgf')
```
